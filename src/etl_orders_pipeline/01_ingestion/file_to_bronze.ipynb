```python

files_df = (
  spark.readStream
  .format("csv")
  .option("header", "true")
  .schema("""
      order_id INT,
      customer_id INT,
      product STRING,
      price DOUBLE,
      quantity INT,
      timestamp STRING
  """)
  .load("/Volumes/workspace/default/etl_volume/orders_input")
)

# When not using Auto Loader, you must define schema manually.
```

```python
files_df.writeStream \
  .format("delta") \
  .trigger(availableNow=True) \
  .option(
      "checkpointLocation",
      "/Volumes/workspace/default/etl_volume/checkpoints/files_bronze"
  ) \
  .toTable("workspace.default.bronze_orders")
```
