```python
from pyspark.sql.functions import sum

silver_df = spark.readStream.table("workspace.default.silver_orders_etl")

gold_df = (
    silver_df
    .groupBy("product")
    .agg(sum("total_amount").alias("total_revenue"))
)
```

```python
gold_df.writeStream \
  .format("delta") \
  .outputMode("complete") \
  .trigger(availableNow=True) \
  .option(
      "checkpointLocation",
      "/Volumes/workspace/default/etl_volume/checkpoints/silver_gold"
  ) \
  .toTable("workspace.default.gold_product_sales_etl")
```
