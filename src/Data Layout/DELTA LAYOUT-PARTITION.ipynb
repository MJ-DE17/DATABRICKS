```python
df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load("/Volumes/workspace/default/my_files/customers-1000.csv")

display(df)
```

```python

df = df.toDF(*[c.replace(" ", "_").lower() for c in df.columns])

display(df)
```

```python
# normal pation table folders
df.write \
.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_files/customers_delta_normal")
```

```python
# Here all data is stored in the same folder.
normal_df = spark.read.format("delta") \
.load("/Volumes/workspace/default/my_files/customers_delta_normal")

display(normal_df)
```

```python
# Write Partitioned Delta Table
df.write \
.format("delta") \
.mode("overwrite") \
.partitionBy("country") \
.save("/Volumes/workspace/default/my_files/customers_delta_partitioned")
```

```python
partitioned_df = spark.read.format("delta") \
.load("/Volumes/workspace/default/my_files/customers_delta_partitioned")

display(partitioned_df)

# Even though data is stored in folders, Spark reads it as one table.
```

```python
# Run Query Using Partition Column
partitioned_df.filter("Country = 'India'")
```

```python
partitioned_df.createOrReplaceTempView("customers_partitioned")
```

```python
%sql

SELECT *
FROM customers_partitioned
WHERE Country = 'India'
```
