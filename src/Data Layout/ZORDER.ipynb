```python
df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load("/Volumes/workspace/default/my_files/customers-1000.csv")


df = df.toDF(*[c.replace(" ", "_").lower() for c in df.columns])

display(df)
```

```python
# CREATE DELTA TABLE

df.write \
.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_files/customers_zorder_demo")
```

```python
%sql
OPTIMIZE delta.`/Volumes/workspace/default/my_files/customers_zorder_demo`
ZORDER BY (Country)
```

```python
# dsiplay uing df

partitioned_df = spark.read.format("delta") \
.load("/Volumes/workspace/default/my_files/customers_zorder_demo")

china_df  = partitioned_df.filter("country = 'China'")
display(china_df )
```

```python
# display using view creation
china_df.createOrReplaceTempView("customers_zorder")
```
