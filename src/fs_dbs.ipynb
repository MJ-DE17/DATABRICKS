```python
# same for move 
# cp is replace with mv
# The correct file to move is fs_demo_copy.txt (previously created in Cell 5)
# Ensure the copy exists before move

# Minimal fix: copy file from original directly in this cell to avoid FileNotFoundError
# If already exists, overwrite is allowed

dbutils.fs.cp("/Volumes/workspace/default/my_files/fs_demo/fs_demo.txt","/Volumes/workspace/default/my_files/fs_demo/fs_demo_copy.txt",True)

dbutils.fs.mv("/Volumes/workspace/default/my_files/fs_demo/fs_demo_copy.txt","/Volumes/workspace/default/my_files/fs_demo/sample_moved.txt")
```
