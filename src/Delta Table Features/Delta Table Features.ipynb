```python
# read

orders_df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load("/Volumes/workspace/default/stream_demo/input/orders_stream_sample1.csv")

display(orders_df)
```

```python
# We store the data as Delta.
orders_df.write.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/stream_demo/output/orders_delta")
```
