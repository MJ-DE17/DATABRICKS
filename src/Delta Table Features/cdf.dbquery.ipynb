```sql
CREATE TABLE cdf_demo (
  id INT,
  name STRING,
  salary INT
)
TBLPROPERTIES (
  delta.enableChangeDataFeed = true
);
-- Step 2 – Insert data
INSERT INTO cdf_demo VALUES
(1, 'Alice', 50000),
(2, 'Bob', 60000);

-- Step 3 – Update data
UPDATE cdf_demo
SET salary = 65000
WHERE id = 2;

-- Step 4 – Delete data
DELETE FROM cdf_demo
WHERE id = 1;

-- Step 5 – Read change data
SELECT *
FROM table_changes('cdf_demo', 0);
```
