```python
# csv

df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load("/Volumes/workspace/default/my_files/customers-1000.csv")

display(df)
```

```python
# json read :

df = spark.read.format("json") \
.load("/Volumes/workspace/default/my_files/customers.json")

display(df)
```

```python
# parquet write 

df.write.format("parquet") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_files/customers_parquet")
```

```python

# delta write

df.write.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_files/customers_delta")
```

```python
# delta read :
df = spark.read.format("delta") \
.load("/Volumes/workspace/default/my_files/customers_delta")

display(df)
```
