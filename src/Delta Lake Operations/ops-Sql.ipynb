```python
%sql
MERGE INTO workspace.default.customers_1000 AS target
USING workspace.default.japan_customers AS source
ON target.`Index` = source.`Index`

WHEN MATCHED THEN
UPDATE SET
target.`Customer Id` = source.`Customer Id`,
target.`First Name` = source.`First Name`,
target.`Last Name` = source.`Last Name`,
target.`Company` = source.`Company`,
target.`City` = source.`City`,
target.`Country` = source.`Country`,
target.`Phone 1` = source.`Phone 1`,
target.`Phone 2` = source.`Phone 2`,
target.`Email` = source.`Email`,
target.`Subscription Date` = source.`Subscription Date`,
target.`Website` = source.`Website`

WHEN NOT MATCHED THEN
INSERT (
`Index`,
`Customer Id`,
`First Name`,
`Last Name`,
`Company`,
`City`,
`Country`,
`Phone 1`,
`Phone 2`,
`Email`,
`Subscription Date`,
`Website`
)
VALUES (
source.`Index`,
source.`Customer Id`,
source.`First Name`,
source.`Last Name`,
source.`Company`,
source.`City`,
source.`Country`,
source.`Phone 1`,
source.`Phone 2`,
source.`Email`,
source.`Subscription Date`,
source.`Website`
);

SELECT *
FROM workspace.default.customers_1000
ORDER BY `Index` DESC
```
