```python
data = [
(1,"Alice","USA",1200),
(2,"Bob","India",900),
(3,"Charlie","USA",1500),
(4,"David","India",700),
(5,"Eva","UK",1300)
]

columns = ["id","name","country","sales"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
df.write.format("delta") \
.mode("overwrite") \
.save("dbfs:/FileStore/medallion/bronze_sales")
```

```python
bronze_df = spark.read.format("delta") \
.load("dbfs:/FileStore/medallion/bronze_sales")

bronze_df.show()
```

```python
from pyspark.sql.functions import sum

gold_df = silver_df.groupBy("country") \
.agg(sum("sales").alias("total_sales"))

gold_df.show()
```

```python
gold_df.write.format("delta") \
.mode("overwrite") \
.save("dbfs:/FileStore/medallion/gold_sales")
```

```python
spark.read.format("delta") \
.load("dbfs:/FileStore/medallion/gold_sales") \
.show()
```

```python
spark.sql("""
CREATE TABLE gold_sales
USING DELTA
LOCATION 'dbfs:/FileStore/medallion/gold_sales'
""")
```
